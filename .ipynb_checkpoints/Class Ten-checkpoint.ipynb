## Handling Null Values

One possible option is to drop rows containing missing values.

```python
df.dropna()
```

```python
df
```

By default, any method which changes our dataframe will return a copy of the changed dataframe keeping the original as it is. To change the frame itself, we supply an option called `inplace=True`

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.shape
```

We can throw away the column that has missing values.

```python
df.dropna(axis=1, inplace=True)
```

```python
df.shape
```

```python
df = pd.read_csv('housing.csv')
```

Quickly comparing columns before and after.

```python
df_cols_dropped = df.dropna(axis=1)
```

```python
df_cols_dropped.columns
```

```python
set(df.columns) - set(df_cols_dropped.columns)
```

Filling missing values with something!

```python
imputed_df = df.fillna(value=df['total_bedrooms'].mean())
```

# Column-wise Operations

```python
df = original_df
```

```python
df.columns
```

```python
df['total_rooms'] / df['population']
```

```python
df['rooms_per_person'] = df['total_rooms'] / df['population']
```

```python
df.columns
```

```python
inland_df = df[df['ocean_proximity'] == 'INLAND']
```

```python
inland_df.shape
```

```python
df2 = df[(df['ocean_proximity'] == 'INLAND') & (df['population'] > 500)]
```

```python
df2.shape
```

```python
df3 = df[(df['ocean_proximity'] == 'INLAND') | (df['population'] > 500)]
```
